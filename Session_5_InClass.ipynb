```python
# 1
import pandas as pd

df = pd.read_csv('./data/homes.csv')
```

```python
#2
print(df.describe())
print(df.dtypes)
print(df.head())
```

From the above functions we can see that the dataset contains primarily integers with one flaot variable. We also can see that there are no missing values in the dataframe. 

```python
# 4
import matplotlib.pyplot as plt

# Histogram
plt.figure(figsize=(10, 6))
df['Sell'].hist(bins=20)
plt.title('Sell Price vs Frequency')
plt.xlabel('Sell Price in Thousands')
plt.ylabel('Frequency')
plt.show()

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df[' "Living"'], df['Sell'], edgecolor='black')
plt.title('Living Area vs Sell Price')
plt.xlabel('Living Area (1000 sqft)')
plt.ylabel('Sell Price ($1000s)')
plt.show()

# Boxplot
plt.figure(figsize=(10, 6))
df[' "Taxes"'].plot(kind='box')
plt.title('Taxes Boxplot')
plt.ylabel('Annual Taxes')
plt.show()
```

```python
# 5
from sklearn.linear_model import LinearRegression

X = df[[' "Taxes"']] 
y = df['Sell']

model = LinearRegression()

model.fit(X, y)

print('Intercept:', model.intercept_)
print('Slope:', model.coef_[0])
print(model.score(X, y))

```

The intercept we found using Taxes as the independent variable and Sell as the dependent variable was 16.87. This shows the estimated sell price when taxes are zero. The slope is 0.04 which shows that for each additional unit of taxes paid, the sell price can be expected to increase on average by 0.04 thousand dollars. The r-squared value is 0.808 which is relatively high, showing taxes is a solid predictor for the sell price. 

```python
# 6

X = df[[' "Living"']] 
y = df['Sell']

model = LinearRegression()

model.fit(X, y)

print('Intercept:', model.intercept_)
print('Slope:', model.coef_[0])
print(model.score(X, y))
```

```python
# Comparison
Our first model using Taxes as the independent variable had a higher r-squared value,
showing that it may be better at modeling the sell prices. On the other hand, the second model
had a better slope, which means it may have a bigger impact on the price. Given the results, the
first model seems to fit better. 
```
